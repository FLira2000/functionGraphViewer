```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def functionCreator(a, b, c, x, aExp, bExp):
    return (pow(a*x, aExp) + pow(b*x, bExp) + c)
```

```python
x = []
```

```python
i = -99
while i < 100:
    x.append(i)
    i+=1
```

```python
y = []
for i in range(-99, 100):
    y.append(functionCreator(1, 0, 0, i, 3, 0))
```

```python
plt.plot(x, y)
plt.show()
```

```python
y = []
for i in range(-99, 100):
    y.append(functionCreator(1, 0, 0, i, 2, 0))
```

```python
plt.plot(x, y)
plt.show()
```
